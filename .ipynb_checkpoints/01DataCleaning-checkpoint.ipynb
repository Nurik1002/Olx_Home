```python
def count_0(series):
    s = 0
    for i in series:
        if i==0:
            s +=1
    return s    

def df_count_0(df, prices_series):
    
    for i in df.columns:
        print(i, " 0 lar soni  : ",  count_0(df[i]))
        print(i, " corr prices : ", df[i].corr(prices_series), end="\n")
```

```python
# umumiy maydoni corr normal
print(count_0(data['Umumiy maydoni']))
data['Umumiy maydoni'].corr(data['Prices'])
```

# Model fit

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
print("R^2 : ", r2_score(y_test, y_pred))
print("MAE :", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```
